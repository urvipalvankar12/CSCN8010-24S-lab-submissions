```python
import matplotlib.pyplot as plt
names= ['Dog','Cat','Monkey','Lion','Tiger']
counting=[5,6,7,8,4]
colors=['Red','Yellow','Pink','grey','gold']
explode= (0.5,0,0,0,0)

fig, ax=plt.subplots()
ax.pie(counting, explode=explode, labels=names, colors=colors, startangle=90)
ax.axis('equal')
ax.set_title('Animal Popularity in zoo')
plt.show()
```

```python

flights = ("Air India, Air canada, Air swiss, Air Ethopia")
plt.bar(names, counting, data=flights, width= 1)
```
