Data Visualization using Matplotlib to display a pie chart on top English league team


```python
import pandas as pd
import matplotlib as mpl
import seaborn as sb
import plotly as pl
```

```python
from matplotlib import pyplot as plt
```

```python
Club=['Man City','Arsenal','Liverpool','Man U','Chelsea']
Points=[88,86,79,57,60]
fig,ax=plt.subplots()
ax.pie(Points,labels=Club,autopct="%1.1f%%",shadow=True, startangle=-90)
ax.axis("equal")
plt.show()
```

Pottling a bar chart rusing matplotlib

```python
color=["skyblue","red","darkred","red","blue"]
plt.bar(Club,Points,color=color,width=0.5)
```
